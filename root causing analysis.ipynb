```python
def collect_failure_info():
    print("=== Equipment Failure Information ===")
    equipment = input("Enter equipment name: ")
    failure_date = input("Enter date of failure (YYYY-MM-DD): ")
    failure_desc = input("Describe the failure: ")
    impact = input("What was the impact of the failure? ")
    return {
        "equipment": equipment,
        "date": failure_date,
        "description": failure_desc,
        "impact": impact
    }

def five_whys():
    print("\n=== 5 Whys Analysis ===")
    reasons = []
    for i in range(1, 6):
        reason = input(f"Why #{i}? ")
        reasons.append(reason)
    root_cause = reasons[-1]
    return reasons, root_cause

def generate_report(info, reasons, root_cause):
    print("\n=== RCA REPORT ===")
    print(f"Equipment: {info['equipment']}")
    print(f"Date of Failure: {info['date']}")
    print(f"Failure Description: {info['description']}")
    print(f"Impact: {info['impact']}\n")

    print("5 Whys:")
    for i, reason in enumerate(reasons, 1):
        print(f"  Why #{i}: {reason}")
    
    print(f"\n>> Root Cause Identified: {root_cause}")
    action = input("\nSuggested Corrective Action: ")
    print("\nCorrective Action Plan:")
    print(f"  - {action}")

def main():
    info = collect_failure_info()
    reasons, root_cause = five_whys()
    generate_report(info, reasons, root_cause)
if  __name__ == "__main__":
    main()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data: Replace with your own failure causes
failure_data = [
    'Motor Overload', 'Bearing Failure', 'Overheating', 'Motor Overload',
    'Bearing Failure', 'Power Supply Issue', 'Motor Overload',
    'Vibration', 'Overheating', 'Bearing Failure',
    'Overheating', 'Power Supply Issue', 'Motor Overload'
]

# Create a DataFrame
df = pd.DataFrame(failure_data, columns=['Failure Cause'])

# Count frequency of each cause
failure_counts = df['Failure Cause'].value_counts()

# Plot 1: Pareto Chart (Sorted Bar Plot)
plt.figure(figsize=(10, 6))
sns.barplot(x=failure_counts.values, y=failure_counts.index, palette='viridis')
plt.title('Pareto Chart of Equipment Failures')
plt.xlabel('Frequency')
plt.ylabel('Failure Cause')
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 2: Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(failure_counts, labels=failure_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Failure Distribution (Pie Chart)')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
